# Machine Learning with Pyhton: Linear Regression With One Variable

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:\\Users\\HP\\Downloads\\canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('abc')
plt.scatter(df.year,df.abc,color='Green',marker='*')
plt.show()
```

```python
new_df=df.drop('year',axis='columns')
new_df
```

```python
price=df.income
price
```

```python
# create Linear regression object
reg=linear_model.LinearRegression()
```

```python
reg.fit(new_df,price)
```

## 1. Predict income of a year2017  

```python
reg.predict([[2017]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
