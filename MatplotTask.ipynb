```python
#Q.1 import matplot
import matplotlib.pyplot as plt
%matplotlib 
import numpy as np
```

```python
#Q.2 plot a basic line chart using plt.plot
x=[1,2,3,4,5]
y=[10,20,30,40,50]
```

```python
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('basic linear chart')
plt.show() 
```

```python
# Q.3 add title and label to axes
plt.plot(x,y)
plt.xlabel('number')
plt.ylabel('value')
plt.title('line chart')
plt.show()
```

```python
#4 chnage line color,style,markers
plt.plot(x,y,color='pink',linestyle='--',marker='o')
plt.xlabel('number')
plt.ylabel('values')
plt.title('line chart with costomization')
plt.show() 
```

```python
#5 plot multiple lines in the same chart
y1=[5,1.5,2.5,3.5,4.5]
```

```python
plt.plot(x,y,color='blue',linestyle='-',marker='o',label='line 1')
plt.plot(x,y1,color='green',linestyle='--',marker='+',label='line 2')
plt.xlabel('number')
plt.ylabel('value')
plt.title('multiple line line chart')
plt.show()
```

```python
#6 add a legend chart
plt.plot(x,y,color='green',linestyle='dotted',marker='*')
plt.xlabel('number')
plt.ylabel('values')
plt.title('legend chart')
plt.legend()
plt.show()
```

```python
#7 save plot as an image file 
plt.plot(x,y,color='yellow',linestyle='dotted',marker='*')
plt.xlabel('number')
plt.ylabel('value')
plt.title ('line chart with image')
plt.legend()
plt.savefig('line_chart.png',dpi=300)
plt.show() 
```

```python
#8 plot bar chart using plt.bar 
categories=['A','B','C'] 
values=[20,11,52,]
```

```python
plt.bar(categories,values,color='orange')
plt.xlabel('categeories')
plt.ylabel('values') 
plt.title('plot bar') 
plt.show()  
```

```python
#9 create histogram
data=[10,12,15,16,18,20]
```

```python
plt.hist(data,bins=5,color='skyblue',edgecolor='green')
plt.title("histogram of data")
plt.xlabel("value ranges")
plt.ylabel("frequency")
plt.show()
```

```python
#10 plot a pie chart
labels=['Banana','Mango','Dates','Cherrie','Apple']
sizes=[25,30,15,18,6]
colors=['yellow','orange','brown','pink','red']
explode=(0,0.1,0,0,0)
plt.pie(sizes,labels=labels,colors=colors,explode=explode,autopct='%1.1f%%',shadow=True,startangle=140)
plt.axis('equal')
plt.title('Fruit Distribution')
plt.show()
```

```python
#11 use subplots to display multiple plots
fig,axs=plt.subplots(1,3, figsize=(15,4))

axs[0].plot(x,y, marker='o',color='blue')
axs[0].set_title('line chart')
axs[0].set_xlabel('x-axis')
axs[0].set_ylabel('y-axis')

axs[1].plot(x,y, marker='o',color='brown')
axs[1].set_title('bar chart')
axs[1].set_xlabel('category')
axs[1].set_ylabel('value')

axs[2].pie(sizes,labels=labels,colors=colors,explode=explode,autopct='%1.1f%%',shadow=True,startangle=140)
axs[2].axis('equal')
axs[2].set_title('Fruit Distribution')
plt.tight_layout()
plt.show()
```

```python
#12 costomize ticks and grid lines
plt.plot(x,y,marker='*',color='purple')
plt.title('costomized ticks and charts')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xticks(ticks=[0,1,2,3,4,5],labels=['zero','one','two','three','four','five'],rotation=45)
plt.yticks(ticks=[0,5,10,15,20,25])
plt.grid(True,linestyle='dotted',linewidth=0.7,color='pink',alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
#13
fig,axs=plt.subplots(2,2,figsize=(12,8))

axs[0,0].plot(x,y)
axs[0,0].set_title('top left')


axs[0,1].bar(x,y)
axs[0,1].set_title('top top right')


axs[1,0].scatter(x,y)
axs[1,0].set_title('bottom left')

axs[1,1].pie([20,30,50],labels=['a','b','c'])
axs[1,1].set_title('bottom right')

plt.tight_layout()
plt.show()
```

```python
#14 plot a scatter plot using plt.scatter
plt.scatter(x,y,color='red',marker='*',s=100,edgecolors='purple')
plt.title('scatter plot example')
plt.xlabel('x values')
plt.ylabel('y label')
plt.grid(True)
plt.show()
```

```python
# 16 
labels = ['A', 'B', 'C']
series1 = [10, 15, 7]
series2 = [8, 9, 13]
series3 = [12, 10, 10]
plt.bar(labels, series1, label='Series 1')
plt.bar(labels, series2, bottom=series1, label='Series 2')
plt.bar(labels, series3, bottom=np.add(series1, series2), label='Series 3')
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Stacked Bar Chart')
plt.legend()
plt.show()
```

```python
#17
fig, ax = plt.subplots()
ax.set_facecolor('#f0f0f0')  
fig.patch.set_facecolor('#f0f0f0')  
ax.bar(labels, series1, label='Series 1')
ax.bar(labels, series2, bottom=series1, label='Series 2')
ax.bar(labels, series3, bottom=np.add(series1, series2), label='Series 3')
ax.set_xlabel('Category', color='blue')
ax.set_ylabel('Value', color='blue')
ax.set_title('Stacked Bar Chart', color='blue')
ax.spines['bottom'].set_color('blue')
ax.spines['left'].set_color('blue')
ax.spines['right'].set_color('blue')
ax.spines['top'].set_color('blue')
ax.tick_params(axis='x', colors='blue')
ax.tick_params(axis='y', colors='blue')
ax.legend(facecolor='#f0f0f0', edgecolor='blue', labelcolor='blue')

plt.show()

```

```python
x = np.arange(1, 11)
y1 = np.random.rand(10) * 10  # Data for left axis
y2 = np.random.rand(10) * 100  # Data for right axis
fig, ax1 = plt.subplots()

ax1.plot(x, y1, color='blue')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Left Axis', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax2 = ax1.twinx()
ax2.plot(x, y2, color='red')
ax2.set_ylabel('Right Axis', color='red')
ax2.tick_params(axis='y', labelcolor='red')
plt.show()

```

```python
x = np.arange(1, 11)
y = np.random.rand(10) * 10
styles = ['ggplot', 'seaborn', 'fivethirtyeight', 'dark_background']
for style in styles:
    plt.style.use(style)
    plt.figure()
    plt.plot(x, y)
    plt.title(f'Style: {style}')
    plt.show()
```
